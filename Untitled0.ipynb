```python
import pandas as pd

# 1. Load the data
df_true = pd.read_csv('True.csv')
df_fake = pd.read_csv('Fake.csv')

# 2. Add labels: 1 for Real, 0 for Fake
df_true['label'] = 1
df_fake['label'] = 0

# 3. Combine them into one single dataframe
df = pd.concat([df_true, df_fake], axis=0).reset_index(drop=True)

# 4. Shuffle the data
# If we don't shuffle, the model will see 20,000 real news stories
# followed by 20,000 fake ones, which ruins the learning process.
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

print(f"Total rows in dataset: {len(df)}")
df.head()
```

```python
# Create a combined column for title and text
df['full_text'] = df['title'] + " " + df['text']

# Keep only the columns we actually need for the model
df = df[['full_text', 'label']]

# Check for any missing values
print("Missing values:\n", df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.countplot(x='label', data=df, palette='viridis')
plt.title('Distribution of Fake (0) vs Real (1) News')
plt.show()

# Print exact percentages
print(df['label'].value_counts(normalize=True) * 100)
```

```python
# Calculate word count for each article
df['word_count'] = df['full_text'].apply(lambda x: len(x.split()))

print(f"Average word count: {df['word_count'].mean():.2f}")
print(f"Max word count: {df['word_count'].max()}")

# Visualize the distribution of text lengths
plt.figure(figsize=(10, 5))
sns.histplot(df['word_count'], bins=50, kde=True)
plt.title('Distribution of Article Word Counts')
plt.axvline(512, color='red', linestyle='--', label='BERT Token Limit')
plt.legend()
plt.show()
```

```python
import torch

class FakeNewsDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

# Create the final datasets
train_dataset = FakeNewsDataset(train_encodings, train_labels)
val_dataset = FakeNewsDataset(val_encodings, val_labels)

print("✅ Day 3 Complete: Datasets are ready for the model!")
```

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer

# Load DistilBERT with 2 output labels (0 and 1)
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, f1_score

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    acc = accuracy_score(labels, predictions)
    f1 = f1_score(labels, predictions, average='binary')
    return {'accuracy': acc, 'f1': f1}
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=2,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    warmup_steps=500,
    weight_decay=0.01,

    # --- FIXED SECTION ---
    logging_dir='./logs',            # Use the original name
    logging_steps=100,
    eval_strategy="steps",           # Updated naming for 2026
    eval_steps=100,
    save_strategy="steps",
    load_best_model_at_end=True,
    # ---------------------

    report_to="none"                 # Avoids external login prompts
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
)

# Start training
trainer.train()
```

```python
import re

def clean_reuters_tag(text):
    # Removes patterns like "WASHINGTON (Reuters) - " or "LONDON (Reuters) - "
    return re.sub(r'^.*?\(Reuters\)\s*-\s*', '', text)

# Apply to the full text
df['full_text_cleaned'] = df['full_text'].apply(clean_reuters_tag)

print("Original:", df['full_text'].iloc[0][:50])
print("Cleaned: ", df['full_text_cleaned'].iloc[0][:50])
```

```python
improved_args = TrainingArguments(
    output_dir='./results_improved',
    num_train_epochs=3,              # Increased to 3 epochs
    learning_rate=1e-5,              # Lowered for precision
    per_device_train_batch_size=8,   # Lowered batch size because max_length is higher (OOM protection)
    per_device_eval_batch_size=8,
    weight_decay=0.01,
    eval_strategy="steps",
    eval_steps=100,
    load_best_model_at_end=True,
    report_to="none"
)

trainer_improved = Trainer(
    model=model,
    args=improved_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
)

trainer_improved.train()
```

```python
!pip install -q gradio
import gradio as gr

def predict_news(text):
    # 1. Clean the text
    text = clean_reuters_tag(text)
    # 2. Tokenize
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512).to("cuda")
    # 3. Predict
    with torch.no_grad():
        logits = model(**inputs).logits
    prediction = torch.argmax(logits, dim=-1).item()

    return "REAL NEWS" if prediction == 1 else "FAKE NEWS"

# Launch the UI
interface = gr.Interface(
    fn=predict_news,
    inputs=gr.Textbox(lines=10, placeholder="Paste a news article here..."),
    outputs="text",
    title="BERT Fake News Detector",
    description="Enter news text to see if the Transformer model thinks it's Real or Fake."
)
interface.launch(share=True) # share=True gives you a public URL for 72 hours!
```
